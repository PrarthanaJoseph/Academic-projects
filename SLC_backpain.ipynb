### Feature Selection using Backward elimination

```python
import statsmodels.api as sm
x=df.drop(['status_LE'],axis=1)
y=df['status_LE']
```

```python
#Backward Elimination
cols = list(x.columns)
pmax = 1
while (len(cols)>0):
    p= []
    X = x[cols]
    xc = sm.add_constant(X)
    model = sm.OLS(y,xc).fit()
    p = pd.Series(model.pvalues.values[1:],index = cols)      
    pmax = max(p)
    feature_with_p_max = p.idxmax()
    if(pmax>0.05):
        cols.remove(feature_with_p_max)
    else:
        break
selected_features_BE = cols

print(model.pvalues)
print('selected features:',selected_features_BE)
```

```python
x_final=df[cols] 
x_final.shape
```

## Logistic regression without standardization

```python
from sklearn.model_selection import train_test_split  # import train_test_split function from sklearn library
```

```python
# using selected features we use logistic regression to fit the model
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_final,y,test_size=0.2,random_state=30) 
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)# train the model
```

```python
y_pred=model.predict(x_test)# predicted y value using x_test
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
###  -- Evaluation metrics : Confusion matrix used for calculating specificity and sensitivity
from sklearn import metrics
from sklearn.metrics import confusion_matrix

```
